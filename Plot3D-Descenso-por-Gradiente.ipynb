```python
import numpy as np
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
fig = pp.figure()
fig.set_size_inches(10,7)

ax = pp.axes(projection = '3d')
x = np.arange(-10.0, 10.0, 0.1)
y = np.arange(-10.0, 10.0, 0.1)
X, Y = np.meshgrid(x, y)

Z = 4.0*np.power(X,2)-4.0*np.dot(X,Y)+2*np.power(Y,2)
#ax.plot_wireframe(X, Y, Z, color = 'green', alpha = 0.43)
ax.plot_surface(X, Y, Z, cmap = 'viridis', edgecolor = 'blue')
ax.set_title('f(x, y, z)')

pp.savefig('Plot3D.png', bbox_inches = 'tight')
```
