## Ejemplo de carga de las imágenes, conversión en escala de grises y visualización

```python
import numpy as np
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
icons = np.load('Icons-50.npy', allow_pickle = True).item()
len(icons)
```

```python
icons['class']
```

```python
icons['image'][0]
```

```python
imagen = icons['image'][0]
print(imagen.shape)
imagen = np.rollaxis(imagen, 0, 3)
print(imagen.shape)
imagen_plot = pp.imshow(imagen)
```

```python
def rgb2gray(imgrgb):
    return np.dot(imgrgb[...,:3], [0.299,0.587,0.144])

gris = rgb2gray(imagen)

imgplotgris = pp.imshow(gris, cmap = 'gray')

print(gris.shape)
```
